```python
# deleting out missing values
import pandas as pd
dht= pd.read_csv('Daily Household Transactions.csv')
new_dht = dht.dropna()    
print(new_dht.to_string())
```

```python
# removing duplicates
new_dht.drop_duplicates(inplace=True)
```

```python
# trend analyis on basis of Daiky transaction 

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 4))
sns.histplot(new_dht['Amount'], bins=2, kde=True)
plt.title('Daily house transaction')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# trend analysis by category
plt.figure(figsize=(36, 18))
sns.countplot(data=new_dht, x='Category', order=new_dht['Category'].value_counts().index)
plt.title('Transaction Counts by Category')

plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# creating pivot table

pivot_table = new_dht.pivot_table(index='Date', columns='Category', values='Amount',
aggfunc='sum', fill_value=0)
```

```python
# Convert 'Date' column to datetime format
new_dht['Date'] = pd.to_datetime(new_dht['Date'], errors='coerce')

# Drop rows with invalid dates
new_dht.dropna(subset=['Date'], inplace=True)

# Resample the data
monthly_data = new_dht.resample('M', on='Date').sum()

plt.figure(figsize=(14, 7))
plt.plot(monthly_data.index, monthly_data['Amount'], marker='o')
plt.title('Monthly Transaction Amounts')
plt.xlabel('Month')
plt.ylabel('Total Amount')

plt.grid(True)
plt.show()
```

```python
# Resample the data to daily frequency
daily_data = new_dht.resample('D', on='Date').sum()

plt.figure(figsize=(14, 7))
plt.plot(daily_data.index, daily_data['Amount'], marker='o')
plt.title('Daily Transaction Amounts')
plt.xlabel('Date')
plt.ylabel('Total Amount')

plt.grid(True)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = pivot_table.corr()

# Plot the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Transaction Categories and Amounts')
plt.show()
```

```python
# Summary of Findings and Visualizations

# 1. Data Cleaning:
# - Missing values were removed from the dataset.
# - Duplicate entries were dropped.
# - The 'Date' column was converted to datetime format and invalid dates were dropped.

# 2. Data Analysis:
# - The dataset contains transactions categorized into various categories such as Food, Transportation, Health, etc.
# - The 'Mode' column indicates the mode of payment used for each transaction.
# - The 'Amount' column represents the transaction amount.

# 3. Visualizations:
# - A histogram of daily household transactions was plotted to show the distribution of transaction amounts.
plt.figure(figsize=(8, 4))
sns.histplot(new_dht['Amount'], bins=20, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# - A count plot of transactions by category was created to show the frequency of transactions in each category.
plt.figure(figsize=(36, 18))
sns.countplot(data=new_dht, x='Category', order=new_dht['Category'].value_counts().index)
plt.title('Transaction Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# - A line plot of monthly transaction amounts was created to show the trend of transactions over time.
plt.figure(figsize=(14, 7))
plt.plot(monthly_data.index, monthly_data['Amount'], marker='o')
plt.title('Monthly Transaction Amounts')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()

# - A line plot of daily transaction amounts was created to show the trend of transactions on a daily basis.
plt.figure(figsize=(14, 7))
plt.plot(daily_data.index, daily_data['Amount'], marker='o')
plt.title('Daily Transaction Amounts')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()

# - A heatmap of the correlation matrix was plotted to show the correlation between different transaction categories.
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Transaction Categories and Amounts')
plt.show()

# Summary:
# - The dataset provides insights into household transactions over time.
# - The visualizations help in understanding the distribution, frequency, and trends of transactions.
# - The correlation matrix indicates the relationships between different transaction categories.
```
