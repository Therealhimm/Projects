```python
# Group by 'Year' and count the number of employees for each year
employees_per_year = salary.groupby('Year').size()

# Plot the number of employees per year
employees_per_year.plot(kind='line', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.title('Number of Employees Per Year')
plt.show()

# Conclude the trend
if employees_per_year.is_monotonic_increasing:
    print("The number of people doing jobs is increasing substantially each year.")
else:
    print("The number of people doing jobs is not increasing substantially each year.")
```

```python
# Convert 'BasePay' to numeric, forcing errors to NaN
salary['BasePay'] = pd.to_numeric(salary['BasePay'], errors='coerce')

# Group by 'Year' and calculate the average 'BasePay'
average_base_pay_per_year = salary.groupby('Year')['BasePay'].mean()

# Plot the average base pay per year
average_base_pay_per_year.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Average Base Pay')
plt.title('Average Base Pay Per Year')
plt.show()
```
